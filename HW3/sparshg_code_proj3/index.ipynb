##  7. VolSDF (15 pts)

We have set alpha and beta parameters as constants instead of learning them (as described in paper: https://arxiv.org/pdf/2106.12052.pdf). Beta is the mean absolute deviation of the laplacian distribution with zero mean of the cumulative distribution function (CDF). Beta controls the smoothening of the surface and alpha is just a density scaling factor. So low alpha means lower density of the final rendering. 

1. As the beta controls smoothness, higher beta would make the surface very smooth, and lower beta would bias the SDF to have sharp boundaries.

2. An SDF would be easier to optimize/train with volume rendering with higher value of beta as smooth surface is more forgiving of missed ray intersections. This means lesser number of rays are needed to make the loss converge to a low value. Also, wider higher beta would cause wider distribution, and that would prevent overfitting.  With lower beta (sharp boundaries), gradients might become unstable.

3. Accuracy would be more with lower value of beta as fine details could be captured due to sharp edges. Even though the loss would easily converge with higher beta value, greater mean deviation would also make it less accurate. So a trade off has to be there between ease of optimization and accuracy of the SDF.
